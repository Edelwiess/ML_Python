```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
def loadmat(path, transpose=True):
    data = sio.loadmat(path)
    X = data['X']
    y = data.get('y')
    y = y.reshape(y.shape[0])  ## 将y变成向量
    
    if transpose:
        X = np.array([im.reshape((20, 20)).T for im in X]) ## 进行转置
        X = np.array([im.reshape(400) for im in X])       ## 再次平化图像以保留向量表示
    
    return X, y
```

```python
X, y = loadmat('ex3data1.mat')
print(X.shape)
print(y.shape)
```

```python
def plot_100_image(X):
    """ sample 100 image and show them
    assume the image is square

    X : (5000, 400)
    """
    size = int(np.sqrt(X.shape[1]))

    # sample 100 image, reshape, reorg it
    sample_idx = np.random.choice(np.arange(X.shape[0]), 100)  # 100*400
    sample_images = X[sample_idx, :]

    fig, ax_array = plt.subplots(nrows=10, ncols=10, sharey=True, sharex=True, figsize=(8, 8))

    for r in range(10):
        for c in range(10):
            ax_array[r, c].matshow(sample_images[10 * r + c].reshape((size, size)),
                                   cmap=matplotlib.cm.binary)
            plt.xticks(np.array([]))
            plt.yticks(np.array([]))
```

```python
plot_100_image(X)
```

## 使用sklearn

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler       ## 进行标准化
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
nn = MLPClassifier(solver='sgd', alpha=1e-05, hidden_layer_sizes=(25, ), random_state=1)
nn.fit(X_train, y_train)
```

```python
y_train_pred = nn.predict(X_train)
```

```python
from sklearn.metrics import classification_report      ## 用来生成评价报告
from sklearn.metrics import confusion_matrix           ## 用来生成混合矩阵
cm_train = confusion_matrix(y_train, y_train_pred)
print(cm_train)
print(classification_report(y_train, y_train_pred))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(nn, X_train, y_train, cv=10)
print(scores)

# rmse_scores = np.sqrt(-scores)
print(scores.mean())
```

```python
x_test = sc.fit_transform(X_test)
y_test_pred = nn.predict(x_test)
```

```python
cm_test = confusion_matrix(y_test, y_test_pred)
print(cm_test)
print(classification_report(y_test, y_test_pred))
```
